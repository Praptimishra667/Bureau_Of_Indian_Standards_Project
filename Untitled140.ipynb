```python
# Install missing libraries if needed
!pip install seaborn openpyxl scikit-learn joblib

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score
import joblib
```

```python
# Upload Excel file from your system
from google.colab import files
uploaded = files.upload()

# Load the "Data" sheet
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name, sheet_name="Data")

# Quick check
print("Shape:", df.shape)
df.head()
```

```python
# Drop imputed fields if present
if "___imputed_fields" in df.columns:
    df = df.drop(columns=["___imputed_fields"])

# Convert purchase date
df["Purchase_Date"] = pd.to_datetime(df["Purchase_Date"], errors="coerce")
df["Purchase_Year"] = df["Purchase_Date"].dt.year
# Standardize text
for c in df.select_dtypes(include="object").columns:
    df[c] = df[c].astype(str).str.strip()

# Normalize compliance
df["Compliance_Status"] = df["Compliance_Status"].str.title()
df["target"] = np.where(df["Compliance_Status"]=="Compliant",1,0)

df.head()
```

```python
# Compliance distribution
sns.countplot(x="Compliance_Status", data=df)
plt.title("Compliance Status Distribution")
plt.show()

# Plant-wise compliance
pd.crosstab(df["Plant"], df["Compliance_Status"]).plot(kind="bar", stacked=True, figsize=(8,5))
plt.title("Compliance by Plant")
plt.show()

# Standard applicability
df["IS_Standard_Applicable"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Standard Applicability")
plt.show()
```

```python
# Date features
df["purchase_year"] = df["Purchase_Date"].dt.year
df["purchase_month"] = df["Purchase_Date"].dt.month
df["component_age_days"] = (pd.Timestamp.today() - df["Purchase_Date"]).dt.days

# Standard applicability flag
df["std_applicable_flag"] = df["IS_Standard_Applicable"].str.lower().map({"yes":1,"no":0}).fillna(0).astype(int)

# Supplier stats
supplier_stats = df.groupby("Supplier")["target"].agg(['count','mean']).rename(columns={"count":"supplier_count","mean":"supplier_compliance_rate"})
df = df.merge(supplier_stats, left_on="Supplier", right_index=True, how="left")

# Standard number length
df["has_standard_number"] = df["Standard_Number"].str.len().fillna(0).gt(0).astype(int)
df["std_num_len"] = df["Standard_Number"].astype(str).str.len()
```

```python
# Features for ML
features = [
    "std_applicable_flag",
    "has_standard_number",
    "std_num_len",
    "supplier_count",
    "supplier_compliance_rate",
    "component_age_days",
    "purchase_year",
    "purchase_month",
    "Plant",
    "Component_Type",
    "Component"
]

X = df[features]
y = df["target"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# Separate numeric & categorical
num_features = X.select_dtypes(include=[np.number]).columns.tolist()
cat_features = X.select_dtypes(include=["object"]).columns.tolist()

# Transformers
num_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="constant", fill_value="__MISSING__")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

# ColumnTransformer
preprocessor = ColumnTransformer([
    ("num", num_transformer, num_features),
    ("cat", cat_transformer, cat_features)
])

# Pipeline with RandomForest
clf = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(n_estimators=200, random_state=42))
])

# Train model
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_proba = clf.predict_proba(X_test)[:,1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Extract feature names
ohe = clf.named_steps["preprocessor"].named_transformers_["cat"].named_steps["onehot"]
cat_names = ohe.get_feature_names_out(cat_features)
feat_names = num_features + list(cat_names)

# Feature importances
importances = clf.named_steps["classifier"].feature_importances_
feat_imp = pd.Series(importances, index=feat_names).sort_values(ascending=False).head(15)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=feat_imp.values, y=feat_imp.index)
plt.title("Top Feature Importances")
plt.show()
```

```python
# Save model
joblib.dump(clf, "BIS_random_forest_pipeline.joblib")

# Predictions dataframe
X_test_copy = X_test.copy()
X_test_copy["Predicted_Proba"] = y_proba
X_test_copy["Prediction"] = y_pred
X_test_copy["Actual"] = y_test

# Save predictions
X_test_copy.to_csv("predictions.csv", index=False)

# Download results
from google.colab import files
files.download("predictions.csv")
files.download("BIS_random_forest_pipeline.joblib")
```
